```python
#saliency weighted semantic network features


import re
rexp=r'[A-Za-z0-9]+'
from math import ceil
count=0
avgsl=0
length=len(s1_train+s2_train)
for sentences in (s1_train+s2_train):
    word_list=re.findall(rexp,sentences)
    count=count+len(word_list)
avgsl=ceil(count/length)




import os
N=0
subdir=os.listdir(path='/media/pranay/New Volume/tweet13/INEXQAcorpus2013')
for folder in subdir:
    val='/media/pranay/New Volume/tweet13/INEXQAcorpus2013'
    val=val+'/'+folder
    N=N+len(os.listdir(val))
    
    
    


def semantic(w,s,model='mikolov_oob'):
    word_list=re.findall(rexp,s)
    if(model=='mikolov_oob'):
        return(max([model_mikolov.similarity(w,words) for words in word_list]))
    elif(model=='word2vec'):
        return(max([model_word2vec_aux.similarity(w,words) for words in word_list]))
    elif(model=='glove840'):
        return(max([model_glove_840b.similarity(w,words) for words in word_list]))
    elif(model=='glove42'):
        return(max([model_glove_42b.similarity(w,words) for words in word_list]))







                              #Inverse document Frequency

def IDF(w):
    count=0
    for folder in subdir:
        val='/media/pranay/New Volume/tweet13/INEXQAcorpus2013'
        val=val+'/'+folder
        files=os.listdir(path=val)
        for xml in files:
            val='/media/pranay/New Volume/tweet13/INEXQAcorpus2013'+'/'+folder+'/'+xml
            with io.open(val,'r',encoding='ascii',errors='ignore') as f:
                content=f.readlines()
                for rows in contents:
                    words=re.findall(rexp,rows)
                if w in words:
                    count+=1                
     
    return(np.log10(N/count))       
        

    

    
    

def saliency_weighted(s1,s2,model_type):
    
    word1=re.findall(rexp,s1)
    word2=re.findall(rexp,s2)
    
        
    if(len(word1)>=len(word2)):
        
        long=s1
        short=s2
        l= word1
        s= word2
    
    else:
        
        long=s2
        short=s1
        l= word2
        s= word1
    
    bin_1,bin_2,bin_3=0,0,0
    s=0
    
    for word in l:
        
        value=0
        a=semantic(word,short,model=model_type)
        value=(IDF(word)*a*2.2)/(a+1.2*(1-0.75+(0.75*len(s)/avgsl)))
        s=s+value
        if(0<=value<0.15):
            bin_1+=1
        elif(0.15<=value<0.4):
            bin_2+=1
        elif(value>=0.4):
            bin_3+=1
            
    return(list((bin_1/s,bin_2/s,bin_3/s)))        
        
        
        
        
```

```python
#Unweighted Semantic network Features


#Feature set 1

def uwsn_feature_1(s1,s2,model_type='mikolov_oob'):
    
    word1=re.findall(rexp,s1)
    word2=re.findall(rexp,s2)
    mat=np.array((len(word1),len(word2)))
    
    if (model_type=='mikolov_oob'):
        for i in range(len(word1)):
            for j in range(len(word2)):
                         
                mat[i][j]=model_mikolov.similarity(word1[i],word2[j])
   
    elif (model_type=='word2vec_aux'):
        for i in range(len(word1)):
            for j in range(len(word2)):
                      
                mat[i][j]=model_word2vec_aux.similarity(word1[i],word2[j])
                
                
    elif (model_type=='glove840'):
        for i in range(len(word1)):
            for j in range(len(word2)):
                      
                mat[i][j]=model_glove_840b.similarity(word1[i],word2[j])
                            
                
    elif (model_type=='glove42'):
        for i in range(len(word1)):
            for j in range(len(word2)):
                      
                mat[i][j]=model_glove_42b.similarity(word1[i],word2[j])
                
    
    bin_1,bin_2,bin_3=0,0,0
    
    for i in range(len(word1)):
        for j in range(len(word2)):
            
            if(-1<=mat[i][j]<0.45):
                bin_1+=1
            elif(0.45<=mat[i][j]<0.8):
                bin_2+=1
            elif(mat[i][j]>=0.8):
                bin_3+=1
     
    return(list((bin_1,bin_2,bin_3)))


#Feature set 2


def uwsn_feature_2(s1,s2,model_type='mikolov_oob'):
    
    word1=re.findall(rexp,s1)
    word2=re.findall(rexp,s2)
    mat=np.array((len(word1),len(word2)))
    
    if (model_type=='mikolov_oob'):
        for i in range(len(word1)):
            for j in range(len(word2)):
                         
                mat[i][j]=model_mikolov.similarity(word1[i],word2[j])
   
    elif (model_type=='word2vec_aux'):
        for i in range(len(word1)):
            for j in range(len(word2)):
                      
                mat[i][j]=model_word2vec_aux.similarity(word1[i],word2[j])
                
                
    elif (model_type=='glove840'):
        for i in range(len(word1)):
            for j in range(len(word2)):
                      
                mat[i][j]=model_glove_840b.similarity(word1[i],word2[j])
                            
                
    elif (model_type=='glove42'):
        for i in range(len(word1)):
            for j in range(len(word2)):
                      
                mat[i][j]=model_glove_42b.similarity(word1[i],word2[j])
                
    
    
    bin_1,bin_2,bin_3=0,0,0
    
    for w in (np.unique(word1)):
        value=0
        
        if w not in (np.unique(word2)):
            value=max(mat[word1.index(w)])
            if(-1<=value<0.45):
                bin_1+=1
            elif(0.45<=value<0.8):
                bin_2+=1
            elif(value>=0.8):
                bin_3+=1
         
        else:
            bin_3+=1
            
    for w in (np.unique(word2)):
        value=0
        
        if w not in (np.unique(word1)):
            value=max(mat[:,word2.index(w)])
            if(-1<=value<0.45):
                bin_1+=1
            elif(0.45<=value<0.8):
                bin_2+=1
            elif(value>=0.8):
                bin_3+=1       
             
    

    return(list((bin_1,bin_2,bin_3)))

    
```

```python
#Text level features 




#Distance between vector means


def distance_features(s1,s2,model_type='mikolov_oob',out_type='cosine'):
    
    word_s1=re.findall(rexp,s1)
    word_s2=re.findall(rexp,s2)
    
    if(model_type=='mikolov_oob'):
        for w1 in word_s1:
            mean_vec_s1=mean_vec_s1+model_mikolov[w1]
        for w2 in word_s2:
             mean_vec_s2=mean_vec_s2+model_mikolov[w2]
    elif(model_type=='word2vec_aux'):
        for w1 in word_s1:
            mean_vec_s1=mean_vec_s1+model_word2vec_aux[w1]
        for w2 in word_s2:
             mean_vec_s2=mean_vec_s2+model_word2vec_aux[w2]
    elif(model_type=='glove840'):
        for w1 in word_s1:
            mean_vec_s1=mean_vec_s1+model_glove_840b[w1]
        for w2 in word_s2:
             mean_vec_s2=mean_vec_s2+model_glove_840b[w2]
    elif(model_type=='glove42'):
        for w1 in word_s1:
            mean_vec_s1=mean_vec_s1+model_glove_42b[w1]
        for w2 in word_s2:
             mean_vec_s2=mean_vec_s2+model_glove_42b[w2] 
    
    mean_vec_s1=mean_vec_s1/len(word_s1)
    mean_vec_s2=mean_vec_s2/len(word_s2)
   
    
    if(out_type=='cosine'):
        
        return(np.dot(mean_vec_s1,mean_vec_s2))
    
    elif(out_type=='euclidean'):
        
        return(np.linalg.norm(mean_vec_s1-mean_vec_s2))
    
    
    

    
    
#Bins of dimensions


def bins_of_dimensions(s1,s2,model_type='mikolov_oob'):
    
    word_s1=re.findall(rexp,s1)
    word_s2=re.findall(rexp,s2)
    
    if(model_type=='mikolov_oob'):
        for w1 in word_s1:
            mean_vec_s1=mean_vec_s1+model_mikolov[w1]
        for w2 in word_s2:
             mean_vec_s2=mean_vec_s2+model_mikolov[w2]
    elif(model_type=='word2vec_aux'):
        for w1 in word_s1:
            mean_vec_s1=mean_vec_s1+model_word2vec_aux[w1]
        for w2 in word_s2:
             mean_vec_s2=mean_vec_s2+model_word2vec_aux[w2]
    elif(model_type=='glove840'):
        for w1 in word_s1:
            mean_vec_s1=mean_vec_s1+model_glove_840b[w1]
        for w2 in word_s2:
             mean_vec_s2=mean_vec_s2+model_glove_840b[w2]
    elif(model_type=='glove42'):
        for w1 in word_s1:
            mean_vec_s1=mean_vec_s1+model_glove_42b[w1]
        for w2 in word_s2:
             mean_vec_s2=mean_vec_s2+model_glove_42b[w2] 
    
    mean_vec_s1=mean_vec_s1/len(word_s1)
    mean_vec_s2=mean_vec_s2/len(word_s2)
    
    bin_1,bin_2,bin_3,bin_4=0,0,0,0
    
    for i in range(len(mean_vec_s1)):
        if(mean_vec_s1[i]<0.001):
            bin_1+=1
        elif(0.001<=mean_vec_s1[i]<0.01):
            bin_2+=1
        elif(0.01<=mean_vec_s1[i]<0.02):
            bin_3+=1
        elif(mean_vec_s1[i]>=0.02):
            bin_4+=1
    
    for i in range(len(mean_vec_s2)):
        if(mean_vec_s2[i]<0.001):
            bin_1+=1
        elif(0.001<=mean_vec_s2[i]<0.01):
            bin_2+=1
        elif(0.01<=mean_vec_s2[i]<0.02):
            bin_3+=1
        elif(mean_vec_s2[i]>=0.02):
            bin_4+=1
            
            
        
    return(list((bin_1,bin_2,bin_3,bin_4)))






    
                
                
                
        
        
    
```

```python
# The Final Train Feature Matrix


num_models=4
feature_train=np.array((len(s1_train),15*num_models))
for i in range(len(s1_train)):
    r=[]
    models=['mikolov_oob','word2vec_aux','glove840','glove42']
    for m in models:
        a=saliency_weighted(s1_train[i],s2_train[i],model_type=m)   # 3 Features
        b=uwsn_feature_1(s1_train[i],s2_train[i],model_type=m)      # 3 Features
        c=uwsn_feature_2(s1_train[i],s2_train[i],model_type=m)      # 3 Features
        d=distance_features(s1_train[i],s2_train[i],model_type=m,out_type='cosine')  #  1 feature
        e=distance_features(s1_train[i],s2_train[i],model_type=m,out_type='euclidean')  #   1 feature
        f=bins_of_dimensions(s1_train[i],s2_train[i],model_type=m)   # 4 features
        r=r+a+b+c+f
        r.append(d)
        r.append(e)
    feature_train[i]=np.array(r)
    

    
# The Final Test Feature Matrix

feature_test=np.array((len(s1_test),15*num_models))
for i in range(len(s1_test)):
    r=[]
    models=['mikolov_oob','word2vec_aux','glove840','glove42']
    for m in models:
        a=saliency_weighted(s1_test[i],s2_test[i],model_type=m)   # 3 Features
        b=uwsn_feature_1(s1_test[i],s2_test[i],model_type=m)      # 3 Features
        c=uwsn_feature_2(s1_test[i],s2_test[i],model_type=m)      # 3 Features
        d=distance_features(s1_test[i],s2_test[i],model_type=m,out_type='cosine')  #  1 feature
        e=distance_features(s1_test[i],s2_test[i],model_type=m,out_type='euclidean')  #   1 feature
        f=bins_of_dimensions(s1_test[i],s2_test[i],model_type=m)   # 4 features
        r=r+a+b+c+f
        r.append(d)
        r.append(e)
    feature_test[i]=np.array(r)


    
```
